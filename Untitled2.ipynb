```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# 読み込むデータが格納されたディレクトリのパス，必要に応じて変更の必要あり
# データの読み込み方がわからない方は、別途配布している、GoogleColabにおけるドライブのマウントの仕方、と、Google Drive上でのフォルダ構造の説明のpdfファイルをご覧ください。
path = "./data/"

df = pd.read_csv(path + 'train.csv')
df_submit = pd.read_csv(path + 'test.csv')
```

```python
missing_list = ['Age', 'Fare', 'Cabin', 'Embarked', 'Name', 'Ticket', 'PassengerId']
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df_submit['Sex'] = df_submit['Sex'].map({'male': 0, 'female': 1})
df.drop(missing_list, axis=1, inplace=True)
df_submit.drop(missing_list, axis=1, inplace=True)
#df.head()
```

```python
sns.heatmap(df.corr(),annot=True,cmap='bwr',linewidths=0.2)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.show()
```

```python
X = df.iloc[:, 1:].values
Y = df.iloc[:, 0].values
X_submit = df_submit.iloc[:, :].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
# X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=42)
lr = LogisticRegression(random_state=42)

lr.fit(X_train,Y_train)
```

```python
lr.score(X_test, Y_test)
```

```python
coefficient = DataFrame()
coefficient["categories"] = DataFrame(df.columns[1:])
coefficient["cofficient"] = DataFrame(lr.coef_[0])
coefficient
```

```python
target = "SibSp"
sns.scatterplot(x = df[target], y = df["Perished"], label = "actual")
sns.scatterplot(x = df[target], y = lr.predict(X), label = "Predict")
```

```python
y_submit = lr.predict(X_submit)
len(X_submit)
```
